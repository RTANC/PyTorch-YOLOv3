```python
import glob
import random
```

```python
files = glob.glob("data/custom/labels/*.txt")
train_files = files[:9664]
test_files = files[9664:]
```

```python
def reset(percent=10):
    return random.randrange(100) < percent
```

```python
count_p = 0
count_n = 0
for line in train_files:
    f = open(line,'r')
    bbs = list(f)
    f.close()
    ll = bbs[0]
    for bb in bbs[4:]:
        c = bb[0]
        if c == "1":
            ll += bb
            count_p += 1
        elif c == "0" and reset():
            ll += bb
            count_n += 1
    print(ll[:-1])
#     break
    f = open(line,'w')
    f.write(ll[:-1])
    f.close()
```

```python
import glob
import random
import os
```

```python
files = glob.glob("data/custom/images/*.png")
print(files)
```

```python
random.shuffle(files)
print(files)
```

```python
NUM_TRAIN = round(len(files) * 0.8)
NUM_VALID = round(len(files) * 0.2)
print(len(files))
print(NUM_TRAIN)
print(NUM_VALID)
training_set = files[:NUM_TRAIN ]
valid_set = files[-NUM_VALID:]
```

```python
print(len(training_set))
print(len(valid_set))
```

```python
file1 = open(os.path.join("data","custom","train.txt"),"w")
for line in training_set:
    file1.write("\n"+line)
file1.close()
```

```python
file1 = open(os.path.join("data","custom","valid.txt"),"w")
for line in valid_set:
    file1.write("\n"+line)
file1.close()
```

```python
files = glob.glob("data/custom/images/*.jpg")
print(len(files))
file1 = open(os.path.join("data","custom","test.txt"),"w")
for line in files:
    file1.write("\n"+line)
file1.close()
```
